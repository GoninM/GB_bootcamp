3. Определите вид функции (чётная / нечётная / общего вида):

$$y=3x+x^3$$

```python
def check_func(val, func):
    if func(val) == func((-1)*val):
        print(f"функция четная")
    elif func(val) == (-1)*func((-1)*val):
        print(f"функция нечетная")
    else:
        print(f"функция общего вида")    
```

```python
def draw_grapg(func, label):
    x = np.linspace(-50, 50, 100)
    y = np.array(func(x))
    
    plt.plot(x,y)
    plt.title(label)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()
```

```python
def func1(x):
    return 3*x+x**3

check_func(3, func1)
```

```python
draw_grapg(func1, "y=3x+x^3")
```

$$y=4^x+4^{-x}$$

```python
def func2(x):
    return 4**x+4**(-x)

check_func(3, func2)
```

```python
draw_grapg(func2, "y=4^(x)+4^(-x)")
```
