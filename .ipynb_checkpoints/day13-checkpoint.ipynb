### matplotlib 


#### Постройте на основе вашего датасета различные типы графиков: 

```python
# гистограммы
n_bins = 50
nrows = 3
ncols=3

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15,15))

axs[0,0].hist(df["Ratings"], bins=n_bins)
axs[0,0].set_title('Ratings')

axs[0,1].hist(df["Price_d"], bins=n_bins)
axs[0,1].set_title('Price_d')

axs[0,2].hist(df["used"], bins=n_bins)
axs[0,2].set_title('used')

axs[1,0].hist(df["Reviews_d"], bins=n_bins)
axs[1,0].set_title('Reviews_d')

axs[1,1].hist(df["price_drop"], bins=n_bins)
axs[1,1].set_title('price_drop')

axs[1,2].hist(df["Mileages_d"], bins=n_bins)
axs[1,2].set_title('Mileages_d')

axs[2,0].hist(df["group"], bins=n_bins)
axs[2,0].set_title('group')

axs[2,1].hist(df["year"], bins=n_bins)
axs[2,1].set_title('year')

plt.show()
```

```python
# ящик с усами, 
n_bins = 50
nrows = 2
ncols=3

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15,15))

axs[0,0].boxplot(df["Price_d"]);
axs[0,1].boxplot(df["Price_d"][df['used']==1]);
axs[0,2].boxplot(df["Price_d"][df['used']==0]);

axs[1,0].boxplot(df["Ratings"]);
axs[1,1].boxplot(df["year"]);
axs[1,2].boxplot(df["Mileages_d"]);

```

```python
df.dtypes!='object'
```

```python
data = df.select_dtypes(exclude='object')
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
data.head()
```

```python
columns=data.columns
columns
```

```python
# тепловая карта
plt.xticks(np.arange(len(columns)), labels=columns, rotation=90)
plt.yticks(np.arange(len(columns)), labels=columns)
plt.title("heatmap")
plt.imshow(data.corr())
plt.colorbar();
```

```python
# диаграммы рассеяния, 
nrows = len(columns)
ncols=len(columns)

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20,20))
fig.subplots_adjust(hspace=0.2, wspace=0.2)

for i in range(len(columns)):
    for j in range(len(columns)):
        axs[j,i].scatter(data[columns[i]], data[columns[j]])
        
        axs[j,i].set(xlabel=columns[i], ylabel=columns[j], title=f"{columns[j]} vs {columns[i]}")

for ax in axs.flat:
    ax.label_outer()    
    
plt.show()
```

```python
nrows = 1
ncols = 1

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(6,6))
fig.subplots_adjust(hspace=0.2, wspace=0.2)

axs.bar(data['year'],data['Price_d'])
    
plt.show()
```

#### seaborn

```python
sns.pairplot(data, hue="used")
```

```python
nrows = len(columns)
ncols = 1

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(18, 15))
fig.subplots_adjust(hspace=1, wspace=0.2)

for i in range(len(columns)):
    sns.histplot(ax=axes[i], data = data[columns[i]],bins=1000)

```

```python
# ящик с усами
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1, figsize=(25,15))
sns.boxplot(x="year", y = "Price_d", hue = "used", data = data, ax=ax1);
sns.boxplot(x="Ratings", y = "Price_d", hue = "used", data = data, ax=ax2);
```

```python
# тепловая карта
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data.corr(), ax=ax)
```

```python
# диаграммы рассеяния, 
nrows = len(columns)
ncols=len(columns)

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20,20))
fig.subplots_adjust(hspace=0.2, wspace=0.2)

for i in range(len(columns)):
    for j in range(len(columns)):
        sns.scatterplot(data=data, x=columns[i], y=columns[j], ax=axs[j,i])
        
        
for ax in axs.flat:
    ax.label_outer()    
    
plt.show()
```
